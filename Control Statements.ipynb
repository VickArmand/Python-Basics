```python
a=20
if a>50:
    print ('In the if body')
print ('Outside the if block')
```

if else statement

Here we specify statements to be executed when the condition is true and also when it is false

```python
i=20
if(i%2==0):
    print('Number is even')
else:
    print('Number is odd')
    
```

Nested if statement

The more the nestings the less efficient the code is hence minimize the nestings

```python
c=25
if c<25:
    if c%2==0:
        print('c is an even number less than 25')
    else:
        print('c is an odd number greater than 25')
        
else:
    print('c is an odd number greater than 25')
```
